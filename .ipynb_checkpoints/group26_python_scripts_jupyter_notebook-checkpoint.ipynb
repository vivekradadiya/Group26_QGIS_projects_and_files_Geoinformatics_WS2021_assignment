### FTP Grab File Function

```python
# This function downloads a file from the FTP directory into the local directory passed as function parameters.
def grabFile(ftpfullname,localfullname):
    try:
        ret = ftp.cwd(".") # A dummy action to check the connection and to provoke an exception if necessary.
        localfile = open(localfullname, 'wb')
        ftp.retrbinary('RETR ' + ftpfullname, localfile.write, 1024)
        localfile.close()
    
    except ftplib.error_perm:
        print("FTP ERROR. Operation not permitted. File not found?")

    except ftplib.error_temp:
        print("FTP ERROR. Timeout.")

    except ConnectionAbortedError:
        print("FTP ERROR. Connection aborted.")

```

### Generate Pandas Dataframe from FTP Directory Listing

```python
import pandas as pd
import os

# This function generates a pandas dataframe from a FTP directory listing.
def gen_df_from_ftp_dir_listing(ftp, ftpdir):
    lines = []
    flist = []
    try:
        # issue the command LIST in the FTP connection 
        res = ftp.retrlines("LIST "+ftpdir, lines.append)
    except:
        print("Error: ftp.retrlines() failed. ftp timeout? Reconnect!")
        return
        
    if len(lines) == 0:
        print("Error: ftp dir is empty")
        return
    
    for line in lines:
        # print(line)
        [ftype, fsize, fname] = [line[0:1], int(line[31:42]), line[56:]]
        
        fext = os.path.splitext(fname)[-1]
        
        if fext == ".zip":
            station_id = int(fname.split("_")[2])
        else:
            station_id = -1 
        
        flist.append([station_id, fname, fext, fsize, ftype])
        
        

    df_ftpdir = pd.DataFrame(flist,columns=["station_id", "name", "ext", "size", "type"])
    return(df_ftpdir)
```

```python
# Generate a pandas dataframe from the FTP directory listing 
df_ftpdir = gen_df_from_ftp_dir_listing(ftp, ftp_dir)
df_ftpdir
```

```python
# extract column names. They are in German (de).
# Use codecs because of difficulties with character encoding (German Umlaute).

import codecs

def station_desc_txt_to_csv(txtfile, csvfile):
    file = codecs.open(txtfile,"r","utf-8")
    r = file.readline()
    file.close()
    colnames_de = r.split()
    colnames_de
    
    # German-English dictionary
    translate = \
    {'Stations_id':'station_id',
     'von_datum':'date_from',
     'bis_datum':'date_to',
     'Stationshoehe':'altitude',
     'geoBreite': 'latitude',
     'geoLaenge': 'longitude',
     'Stationsname':'name',
     'Bundesland':'state'}
    
    colnames_en = [translate[h] for h in colnames_de]
    
    # Skip the first two rows and set the column names.
    df = pd.read_fwf(txtfile,skiprows=2,names=colnames_en, parse_dates=["date_from","date_to"],index_col = 0)
    
    # write CSV file with field separator semicolon
    df.to_csv(csvfile, sep = ";")
    return(df)
```

```python
# Saves stations description in a .csv file and returns a dataframe of it.
basename = os.path.splitext(station_fname)[0]
df_stations = station_desc_txt_to_csv(local_station_dir + station_fname, local_station_dir + basename + ".csv")
df_stations.head()
```

### Join (Merge) the Time Series Columns

https://medium.com/@chaimgluck1/working-with-pandas-fixing-messy-column-names-42a54a6659cd


```python
from datetime import datetime

def prec_ts_to_df(fname):
    
    dateparse = lambda dates: [datetime.strptime(str(d), '%Y%m%d%H') for d in dates]

    df = pd.read_csv(fname, delimiter=";", encoding="utf8", index_col="MESS_DATUM", parse_dates = ["MESS_DATUM"], date_parser = dateparse, na_values = [-999.0, -999])

    #df = pd.read_csv(fname, delimiter=";", encoding="iso8859_2",\
    #             index_col="MESS_DATUM", parse_dates = ["MESS_DATUM"], date_parser = dateparse)
    
    # https://medium.com/@chaimgluck1/working-with-pandas-fixing-messy-column-names-42a54a6659cd

    # Column headers: remove leading blanks (strip), replace " " with "_", and convert to lower case.
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
    df.index.name = df.index.name.strip().lower().replace(' ', '_').replace('(', '').replace(')', '')
    return(df)
```

```python
from zipfile import ZipFile
```

```python
# PRECIPITATION
def prec_ts_merge():
    # Very compact code.
    df = pd.DataFrame()
    for elt in local_zip_list:
        ffname = local_ts_dir + elt
        print("Zip archive: " + ffname)
        with ZipFile(ffname) as myzip:
            # read the time series data from the file starting with "produkt"
            prodfilename = [elt for elt in myzip.namelist() if elt.split("_")[0]=="produkt"][0] 
            print("Extract product file: %s" % prodfilename)
            print()
            with myzip.open(prodfilename) as myfile:
                dftmp = prec_ts_to_df(myfile)
                s = dftmp["r1"].rename(dftmp["stations_id"][0]).to_frame()
                # outer merge.
                df = pd.merge(df, s, left_index=True, right_index=True, how='outer')

    #df.index.names = ["year"]
    df.index.rename(name = "time", inplace = True)
    return(df)
```

## Daily precipitation and cumulative precipitation for all stations located in the counties Olpe and Hochsauerlandkreis

This code section prepares a pandas dataframe to calculate the daily precipitation and cumulative precipitation for all stations located in the counties Olpe and Hochsauerlandkreis to create a bar graph and line graph for each station.

```python
# Extract dataframe for all stations located in the counties Olpe and Hochsauerlandkreis using the station id to prepare a bar graph and a line graph for each station.
df_prec_investigation_olpe_hsk = df_prec_investigation[[216, 1300, 2483, 2947, 3215, 4488, 5468, 6264, 7330]].copy()
df_prec_investigation_olpe_hsk
```

```python
# Add new column 'date' by formatting it to a date only, to groupby all columns from hourly data to daily data in the dataframe.
arr_hourly_index = df_prec_investigation_olpe_hsk.index
df_prec_investigation_olpe_hsk['date'] = arr_hourly_index
df_prec_investigation_olpe_hsk['date'] = df_prec_investigation_olpe_hsk['date'].apply(lambda x: x.date())
df_prec_investigation_olpe_hsk[:25]
```

```python
# Convert hourly precipitation value into daily precipitation rate for all stations located in the counties Olpe and Hochsauerlandkreis by aggregating the precipitation value.
df_prec_investigation_olpe_hsk = df_prec_investigation_olpe_hsk.groupby("date", as_index = False).sum()
df_prec_investigation_olpe_hsk[:60]
```

```python
import datetime

# Create a new dataframe to store daily precipitation sum and cumulative precipitation data of each day from 2017-04-16 to 2017-08-16 for a particular station.
df_olpe_hsk = pd.DataFrame()

for column in df_prec_investigation_olpe_hsk.columns[1:]:
    # Temporary dataframe to add a station's data.
    df = pd.DataFrame()
    df['date'] = df_prec_investigation_olpe_hsk[df_prec_investigation_olpe_hsk.columns[0]].apply(lambda x: x.strftime('%Y-%m-%d'))
    df['station_id'] = column
    df['daily_preci_sum'] = df_prec_investigation_olpe_hsk[column]
    df['cum_preci'] = df.groupby(["station_id"]).cumsum()
    # Append a dataframe for a single station to the main dataframe.
    df_olpe_hsk = df_olpe_hsk.append(df)
    
df_olpe_hsk.reset_index(inplace = True, drop = True)
df_olpe_hsk
```

```python
# calculate the Pearson's correlation between two variables
from numpy.random import randn
from numpy.random import seed
from scipy.stats import pearsonr
from scipy.stats import spearmanr
from scipy.stats import linregress

# seed random number generator
seed(1)

# prepare data
cum_precipitation = df_cum_preci_altitude_coi['cum_preci']
altitude = df_cum_preci_altitude_coi['altitude']

# calculate Pearson's correlation
corr_pearson, _ = pearsonr(cum_precipitation, altitude)
print('Pearsons correlation: %.3f' % corr_pearson)
slope, intercept, r, p, stderr = linregress(cum_precipitation, altitude)
line = f'Regression line: y={intercept:.2f}+{slope:.2f}x, r={r:.2f}'
line

# calculate Spearman's correlation
corr_spearman, _ = spearmanr(cum_precipitation, altitude)
print('Spearmans correlation: %.3f' % corr_spearman)
line
```

## Geopandas - Create a Geo Data Frame

A Geopandas geo data frame is a Pandas data frame enriched with an additional geometry column. Each row in the data frame becomes a location information. Thus a geo-df contains geometry and attributes, i.e. full features. The geo-df is self-contained and complete. It can be easily saved in different vectore file formats, i.e. shapefile or geopackage.

```python
import pandas as pd
from geopandas import GeoDataFrame
from shapely.geometry import Point
import fiona
from pyproj import CRS

#df = pd.read_csv('data.csv')
df = dfNRW

geometry = [Point(xy) for xy in zip(df.longitude, df.latitude)]
crs = CRS("epsg:4326") #http://www.spatialreference.org/ref/epsg/2263/
stations_gdf = GeoDataFrame(df, crs=crs, geometry=geometry)

stations_gdf.head(5)
```

```python
# Filter to extract PRECIPITATION from recent and historical RR product files.

def prec_ts_to_df(fname):

    import pandas as pd
    import pytz
    from datetime import datetime
    
    dateparse = lambda dates: [datetime.strptime(str(d), '%Y%m%d%H') for d in dates]

    df = pd.read_csv(fname, delimiter=";", encoding="utf8", index_col="MESS_DATUM", parse_dates = ["MESS_DATUM"], date_parser = dateparse, na_values = [-999.0, -999])

    #df = pd.read_csv(fname, delimiter=";", encoding="iso8859_2",\
    #             index_col="MESS_DATUM", parse_dates = ["MESS_DATUM"], date_parser = dateparse)
    
    # https://medium.com/@chaimgluck1/working-with-pandas-fixing-messy-column-names-42a54a6659cd

    # Column headers: remove leading blanks (strip), replace " " with "_", and convert to lower case.
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
    df.index.name = df.index.name.strip().lower().replace(' ', '_').replace('(', '').replace(')', '')
    
    # TIME ZONES: https://stackoverflow.com/questions/22800079/converting-time-zone-pandas-dataframe
    # DWD Prec Data is given in UTC
    
    df.index = df.index.tz_localize(pytz.utc)
    
    return(df)
```

```python

csvfname = "prec_ts_appended_3_cols.csv"

first = False

for elt in local_zip_list:
    ffname = local_ts_dir + elt
    print("Zip archive: " + ffname)
    with ZipFile(ffname) as myzip:
        # read the time series data from the file starting with "produkt"
        prodfilename = [elt for elt in myzip.namelist() if elt.split("_")[0]=="produkt"][0] 
        print("Extract product file: %s" % prodfilename)
        print()
        with myzip.open(prodfilename) as myfile:
            dftmp = prec_ts_to_df(myfile)[["stations_id","r1"]]
            # df.rename(columns={'oldName1': 'newName1', 'oldName2': 'newName2'}, inplace=True)
            dftmp.rename(columns={'stations_id': 'station_id', 'r1': 'val', 'mess_datum': 'ts'}, inplace = True)
            dftmp.rename_axis('ts', inplace = True)
            # dftmp.to_csv(f, header=f.tell()==0)
            if (first):
                first = False
                dftmp.to_csv(csvfname, mode = "w", header = True)
            else:
                dftmp.to_csv(csvfname, mode = "a", header = False)
                
```
